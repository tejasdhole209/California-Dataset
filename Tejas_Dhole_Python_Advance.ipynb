**1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.**



```python
# Calculate average median income
dataset = pd.read_csv("/content/sample_data/california_housing_train.csv")
average_median_income = dataset['median_income'].mean()

# Plot distribution of median income
plt.figure(figsize=(10, 6))
sns.histplot(dataset['median_income'], bins=30, color='red')
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()

# Explanation
print(f"Average Median Income: {average_median_income}")
print("Explanation: The histogram shows the distribution of median income values. It appears to be right-skewed, with a peak around the average median income value.")
```

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
# Create a new column 'total_bedroom_size'
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'Small'
    elif 11 <= total_bedrooms < 1000:
        return 'Medium'
    else:
        return 'Large'

dataset['total_bedroom_size'] = dataset['total_bedrooms'].apply(categorize_bedroom_size)
print("Observation: We can see at last extra column is added named= total_bedroom_size")
dataset.head()
```
